```python
import syft as sy
duet = sy.launch_duet(loopback=True)
```

```python
import os
import torchvision
from torchvision import datasets, transforms, models 
from torchvision.datasets import ImageFolder
from torchvision.transforms import ToTensor
from torchvision.utils import make_grid
from torch.utils.data import random_split
from torch.utils.data.dataloader import DataLoader

train_dir = '/scratch/mg6827/data/xray_data/train_split/0/'
classes = os.listdir(train_dir)

train_transform=transforms.Compose([
        transforms.RandomRotation(10),      # rotate +/- 10 degrees
        transforms.RandomHorizontalFlip(),  # reverse 50% of images
        transforms.Resize(100),             # resize shortest side
        transforms.CenterCrop(100),         # crop longest side
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406],
                             [0.229, 0.224, 0.225])
])

trainset = ImageFolder(train_dir, transform=train_transform)
# train_loader = DataLoader(dataset = trainset, batch_size = 1, shuffle = True, num_workers = 4)
# print(train_loader.dataset)
# print(trainset.type)
```

```python
import torch
test_data = []
target = []
for i, data in enumerate(trainset):
#     if i > 30:
#         break
#     print(i)
    #   print(data[0].shape, data[1])
    tmp = data[0].reshape(1, data[0].shape[0], data[0].shape[1], data[0].shape[1])
#     print(tmp.shape)
    test_data.append(tmp)
#     print('data',data[1], torch.FloatTensor(data[1]))
    target.append(data[1])
   
test_data = torch.cat(test_data)
# target = torch.cat(target)
target = torch.LongTensor(target)
# print(target)
```

```python
print(target)
```

```python

test_data = test_data.tag("x_train")
test_data = test_data.describe("Dataset from client 1")
data_ptr = test_data.send(duet, searchable=True)

# target = torch.FloatTensor(target)
target = target.tag("y_train")
target = target.describe("target tags from client 1")
target_ptr = target.send(duet, searchable=True)

train_num = sy.lib.python.Int(test_data.shape[0])
print(train_num)
train_num.tag("train_num")
train_num.send(duet, searchable=True)

duet.requests.add_handler(
    action="accept",
    print_local=True,  # print the result in your notebook
)
```
